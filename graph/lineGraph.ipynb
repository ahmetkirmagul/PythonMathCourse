```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#a bol line
p1 = [-3,-1]
p2 = [4,4]

plt.plot([p1[0],p2[0]],[p1[1],p2[1]],color=[0.9,0.1,0.4],linewidth=5) #[point 1 x, point 1 y],[point 2 x, point 2 y], color, linewidth
```

```python
# a line and a point
x = 3
y= 5
plt.plot(x,y,"ro")
plt.plot([0,x],[0,y],'r') #red y=ax+b

plt.axis("square")
plt.axis([-6,6,-6,6]) #axis
plt.grid()

plt.plot([-6,6],[0,0],'k') #straight black line

plt.show()
```

```python
#y=(|x|)^1/2

x = range(-20,21)

for xi in x:
    plt.plot([0,xi],[0,abs(xi)**(1/2)],color="black")

plt.show()
```

```python
#A square

plt.plot([2,4],[2,2],"r") # 0 ile 2 arası x, 2 y fix
plt.plot([2,4],[4,4],"k") # 0 ile 2 arası x, 0 y fix
plt.plot([2,2],[2,4],"g") # 0 x fix, 0 ile 2 arası y
plt.plot([4,4],[2,4],"m") # 2 x fix, 0 ile 2 arası y

plt.axis('square')
plt.axis([0,6,0,6])
plt.grid()
plt.title("Colorful Square")

plt.show()
```

```python
#Line in slope-intercept form

# y = mx + b

x = [-5,5]
m=3
b=2

y = [340,2345]
for i in range(0,len(x)):
    y[i] = m*x[i] + b

plt.plot(0,b,"o")

plt.plot(x,y, label='y=%sx+%s' %(m,b))
plt.legend()

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')

plt.title("The plot.")

plt.show()
```
