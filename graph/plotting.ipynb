```python
x= [0,10,20,30,40,50,60,70,80]
y= [0,51.18, 50.43, 50.13, 50.70, 51.88, 51.02, 53.62, 54.05]

for i in range(0,len(x)):
    plt.plot(x[i],y[i],"o", label="t%s"%i)#o = circle

plt.grid()#ızgara
plt.legend()#table that shows labels

plt.show()
```

```python
plt.plot(4,2,'r+') #r+ stands for red + sign
axis= plt.gca() #get current axis

print(axis.get_ylim())#y axis'inin limitlerini veriyor
axis.set_ylim([axis.get_ylim()[0],5])

plt.show()
```

```python
import sympy as sym
x = sym.symbols('x')
y = x**2 - 3*x

for pnti in range(-5,6):
    plt.plot(pnti,y.subs(x,pnti),'o')

plt.show()
```
